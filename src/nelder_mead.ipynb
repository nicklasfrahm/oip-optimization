```python
# Import the dependencies.
import sympy as sp
import numpy as np
import matplotlib as mpl
```

```python
# Initialize sympy plotting.
sp.interactive.printing.init_printing(use_unicode=False, wrap_line=False, no_global=True)
%matplotlib inline
```

```python
# Define the function.
x, y = sp.symbols("x y")
f = x * (x + 1) + y**2
display(f)
```

```python
# Plot the function.
yMin = -5
yMax = 5
xMin = -5
xMax = 5
plot = sp.plotting.plot3d(f, (x, xMin, xMax), (y, yMin, yMax));
```

```python
# Plot the function with matplotlib.
npFunc = sp.lambdify((x, y), f, 'numpy')
xAxis = np.linspace(xMin, xMax, 1000)
yAxis = np.linspace(yMin, yMax, 1000)
xGrid, yGrid = np.meshgrid(xAxis, yAxis)

fig = mpl.pyplot.figure()
ax = fig.gca()
con = ax.contourf(xGrid, yGrid, npFunc(xGrid, yGrid), 20, cmap=mpl.cm.jet)
```
